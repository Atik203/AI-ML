```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python

# ^ Create a Simple Dataset
hours = np.array([2,4,6,8,10]).reshape(-1, 1)
marks = np.array([50,55,65,70,75])

hours,marks 
```

```python

 # ^ Train a LInear Regression Model
 model = LinearRegression()
 model.fit(hours, marks)
```

```python

# ^ View the learning parameters: y = mx+c
print("Coefficient (m): ", model.coef_[0])
print("Intercept (c): ", model.intercept_)
```

```python

# ^ Visualize the results
plt.scatter(hours, marks, color='blue', label='Actual Data')
predicted_marks = model.predict(hours)
plt.plot(hours, predicted_marks, color='red', label='Regression Line')  
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.title('Hours Studied vs Marks Obtained')
plt.legend()
plt.grid()
plt.show()
```

```python

# ^ Predict marks for a new student
new_hours = np.array([[7]])
predicted_marks_new = model.predict(new_hours)
print(f"Predicted marks for studying {new_hours[0][0]} hours: {predicted_marks_new[0]}")
```

```python

# ^ Compute the cost (Mean Squared Error)
mse = mean_squared_error(marks, predicted_marks)
print(f"Mean Squared Error on training data: {mse}")
```

```python

# ^ Use a gradient descent based model SGDRegressor
sgd = SGDRegressor(max_iter=10000, learning_rate='invscaling', eta0=0.01,random_state=42)

sgd.fit(hours, marks)
sgd_predicted_marks = sgd.predict(hours)
sgd_mse = mean_squared_error(marks, sgd_predicted_marks)
print(f"SGD Regressor Coefficient (m): {sgd.coef_[0]}")
print(f"SGD Regressor Intercept (c): {sgd.intercept_[0]}")
print(f"Mean Squared Error for SGD Regressor: {sgd_mse}")

```

```python

# ^ Compare both visually

plt.scatter(hours, marks, color='blue', label='Actual Data')
plt.plot(hours, predicted_marks, color='red', label='Linear Regression Line')
plt.plot(hours, sgd_predicted_marks, color='green', linestyle='--', label='SGD Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Obtained')
plt.title('Hours Studied vs Marks Obtained')
plt.legend()
plt.grid()
plt.show()
```
