```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
student = pd.read_csv('sns_data.csv')

student.head()
```

```python
sns.lineplot(data=student,x='study_hours',y='test_score',errorbar=None,hue='gender')
```

```python
sns.relplot(data=student,x='study_hours',y='test_score',hue='gender',kind='line',errorbar=None)
```

```python
# scatter plot using seaborn
sns.scatterplot(data=student,x='study_hours',y='test_score',hue='gender',style='class_level',size='Tshirt_size')
```

```python
tips_data = sns.load_dataset('tips')

tips_data.head()
```

```python
sns.relplot(data=tips_data,x='total_bill',y='tip',hue='sex',kind='scatter')
```

```python
student = pd.read_csv('student_dataset_complete.csv')
student.head()
```

```python
# facet plot using seaborn
sns.relplot(data=student,x='study_hours',y='test_score',col='gender',kind='scatter',row='hostel')
```

```python
sns.relplot(
    data=student,
    x="study_hours",
    y="test_score",
    col="week",
    kind="scatter",
    col_wrap=3
)
```

```python
# histogram using seaborn
sns.histplot(data=student,x='test_score',bins=10,hue='gender',multiple='stack') # multiple = 'dodge', 'stack', 'fill'
```

```python
# kde plot using seaborn
sns.kdeplot(data=tips_data,x='total_bill',hue='sex',fill=True,alpha=0.4)
```

```python
# count and bar plot using seaborn
sns.countplot(data=tips_data,x='day',hue='sex')
```

```python
sns.barplot(data=tips_data, x="day", y="total_bill", hue="sex",errorbar=None)
```

```python
# regplot using seaborn
sns.regplot(data=student,x='study_hours',y='test_score',scatter_kws={'color':'blue'},line_kws={'color':'red'})
```

```python
sns.lmplot(data=student,x='study_hours',y='test_score')
```

```python
#  pair and  join plot using seaborn
sns.pairplot(data=student,hue='gender',diag_kind='kde')
```

```python
sns.jointplot(data=student,x='study_hours',y='test_score',kind='hist',hue='gender')
```

```python
# scatter using plotly
fig = px.scatter(student,x='study_hours',y='test_score',color='gender',title='Study Hours vs Test Score')
fig.show()
```
