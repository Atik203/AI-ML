```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2d line plot
day = np.arange(1, 11)
temp = [30, 32, 34, 33, 31, 29, 28, 27, 26, 25]

plt.plot(day, temp, linestyle='-', color='r',marker='o')
plt.grid(True)
plt.xlabel('Day')
plt.ylabel('Temperature (Â°C)')
plt.title('Daily Temperature Over 10 Days')
```

```python
df = pd.read_csv('enrollment_data.csv')

df.head()

plt.plot(df["Year"], df["Programming"], marker="o", linestyle="-", color="b",label="Programming")
plt.plot(df["Year"], df["Digital Marketing"], marker="s", linestyle="--", color="g",label="Digital Marketing")
plt.xlabel('Year')
plt.ylabel('Number of Enrollments')
plt.legend()
plt.show()
```

```python
plt.scatter(df["Year"], df["Programming"], color="m", marker="^")
plt.scatter(df["Year"], df["Digital Marketing"], color="c", marker="x")
plt.xlabel('Year')
plt.ylabel('Number of Enrollments')
plt.legend(['Programming', 'Digital Marketing'])
plt.show()
```

```python
students = pd.read_csv('student_IQdata.csv')

students.head()
```

```python
plt.scatter(students['Study_Hour'],students['IQ_Score'], color='b', marker='o')

plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Study Hours vs IQ Score')
plt.grid(True)
plt.show()
```

```python
# histogram

plt.hist(students['IQ_Score'], bins=10, color='g', edgecolor='black')
plt.xlabel('IQ Score')
plt.ylabel('Number of Students')
plt.title('Distribution of IQ Scores')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
